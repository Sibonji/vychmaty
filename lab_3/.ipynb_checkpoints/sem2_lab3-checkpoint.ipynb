# Решение дифференциального уравнения методом стрельбы, задача XI 9.3(б)
$$
\begin{cases}
 y'' - x \sqrt y = 0,  &  0 ≤ x ≤ 1\\ 
 y(0) = 0 \\
 \int\limits^1_0 y(x) dx = 1
\end{cases}
$$

Подключим нужные библиотеки

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

Зададим функцию из задачи

```python
def my_func(x, der):
    y = der[0]
    a = der[1]
    return np.array([a, x * math.sqrt(y)])
```

Используем метод Рунге-Кутты 4 порядка:

```python
def runge_kutta_4(y, x, h, func, i):
    k_1 = func(x[i], y[i])
    k_2 = func(x[i] + h / 2, y[i] + k_1 * h / 2)
    k_3 = func(x[i] + h / 2, y[i] + k_2 * h / 2)
    k_4 = func(x[i] + h, y[i] + k_3 * h)

    return h  * (k_1 + 2 * k_2 + 2 * k_3 + k_4) / 6
```

Для  удобства сразу запишем функцию использующую метод рунге Кутты для решения системы

```python
def solve_kutta_4(x, start_vector, h, number_move, func):                      
    y = np.zeros((number_move, np.shape(start_vector)[0]))
    y[0] = start_vector
    sum_integral = 0   
    
    for i in range(0, number_move - 1):
        res = runge_kutta_4(y, x, h, func, i)

        y[i + 1] = y[i] + res
    
        if (i != 0):
            sum_integral += (y[i][0] + y[i - 1][0]) / 2 * h
        
    return sum_integral, y  
```

Функция для отрисовки графика

```python
def draw_plot(x, y):
    plt.figure(figsize = (6, 6))
    y, _ = y.T
    plt.plot(x, y)
    plt.ticklabel_format(style='plain')
    plt.minorticks_on()
    plt.grid(which='major', color='grey', linestyle='dotted')
    plt.xlabel(r'$x$', fontsize=20)
    plt.ylabel(r'$y$', fontsize=20)
    plt.show()
```

Зададим начальные значения и шаг вычисления

```python
x_0_integ = 0
x_1_integ = 1
h = 1e-3
```

Решим при помощи нашей функции и построим график

```python
number_move = int((x_1_integ - x_0_integ) / h) + 1
x = np.linspace(x_0_integ, x_1_integ, number_move)
integral = 0
y = np.empty((0), float)
epsilon = 1e-4
x_0 = 0
y_0 = 0
a_0 = 1.9

while(True):
    start_vector = np.array([y_0, a_0])
    integral, y = solve_kutta_4(x, start_vector, h, number_move, my_func)
    a_0 += h 
    if(abs(integral - 1.0) < epsilon):
        print('Integral: ', integral)
        draw_plot(x, y)
        break
```
